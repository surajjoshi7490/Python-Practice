**6. To find a consecutive pattern like 3,3**  
```python
has_33 = [1, 2, 3, 3]
found = False
for i in range(len(has_33) - 1):
    if has_33[i] == 3 and has_33[i + 1] == 3:
        found = True
        break
print(found)
```

**8. Count Prime Numbers up to n**  
```python
n = 100
primes = []
for number in range(2, n + 1):
    for divider in range(2, number):
        if number % divider == 0:
            break
    else:
        primes.append(number)
print(primes)
print("Total primes:", len(primes))
```
